## Subtitles tokenization

```python
import nltk
import unidecode
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.feature_extraction.text import CountVectorizer

import nltk
nltk.download('rslp')
```

```python
def clean(documents):
   tokenizer = nltk.RegexpTokenizer(r'\w+')
   stopwords = set(nltk.corpus.stopwords.words('portuguese'))
   stemmer = nltk.stem.RSLPStemmer()

   docs = []
   stemmedWords = {}

   for doc in documents:
       doc = doc.lower()
       new_doc = []
       for term in tokenizer.tokenize(doc):
           if term not in stopwords and term != '':
               stemmed = unidecode.unidecode(stemmer.stem(term))
               if stemmed and not stemmed.isdigit():
                   if stemmed in stemmedWords:
                        new_doc.append(stemmedWords[stemmed])
                   else:
                    stemmedWords[stemmed] = term
                    new_doc.append(unidecode.unidecode(stemmedWords[stemmed]))
       if new_doc:
           docs.append(' '.join(new_doc))

   return docs
```

```python
dataset = clean(subtitles)
```

### Running corex topic modeling

```python
from sklearn.feature_extraction.text import CountVectorizer
import scipy.sparse as ss
from corextopic import corextopic as ct
from corextopic import vis_topic as vt # jupyter notebooks will complain matplotlib is being loaded twice
%matplotlib inline
import numpy as np
```

```python
vectorizer = CountVectorizer(binary=True)
wordCount = vectorizer.fit_transform(dataset)
```

```python
# Get words that label the columns (needed to extract readable topics and make anchoring easier)
words = list(np.asarray(vectorizer.get_feature_names()))
```

```python
not_digit_inds = [ind for ind,word in enumerate(words) if not word.isdigit()]
wordCount = wordCount[:,not_digit_inds]
words    = [word for ind,word in enumerate(words) if not word.isdigit()]

wordCount.shape # n_docs x m_words
```

```python
# Train the CorEx topic model with 50 topics
num_topics = 9
topic_model = ct.Corex(n_hidden=num_topics, words=words, max_iter=200, verbose=False, seed=1)
topic_model.fit(wordCount, words=words);
```

```python
# Print a single topic from CorEx topic model
topic_model.get_topics(topic=0, n_words=12)
```

```python
# Print all topics from the CorEx topic model
topics = topic_model.get_topics()
for n,topic in enumerate(topics):
    topic_words,_,_ = zip(*topic)
    print('{}: '.format(n) + ', '.join(topic_words))
```
